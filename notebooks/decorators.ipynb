```python
from datetime import datetime
def now():
    print(datetime.now().date())
f = now
print(f"now对象的值：{now}")
print(f"f  对象的值：{f}")

print(f"now对象的名称：{now.__name__}")
print(f"f  对象的名称：{f.__name__}")

now()
f()
```

```python
from datetime import datetime

def log(func):
    def wrapper(*args, **kw):
        print(f"call {func.__name__}:",end=" ")
        return func(*args, **kw)
    return wrapper

#把@log放到now()函数的定义处，相当于执行了语句：now = log(now)
@log
def now():
    print(datetime.now().date())

print(f"now 对象的名称：{now.__name__}")
now()
# wrapper()在log()函数之内，外部是不能直接访问的
# wrapper()

```

```python
import functools

def log(func):
    @functools.wraps(func)
    def wrapper(*args, **kw):
        print(f"call {func.__name__}:",end=" ")
        return func(*args, **kw)
    return wrapper

@log
def now():
    print(datetime.now().date())

print(f"now 对象的名称：{now.__name__}")
now()
```

```python
from datetime import datetime

def log(text):
    def decorator(func):
        def wrapper(*args, **kw):
            print(f"{text} calling {func.__name__}()")
            return func(*args, **kw)
        return wrapper
    return decorator

@log('execute')
def now():
    print(datetime.now().date())

now()
print(f"now对象的名称：{now.__name__}")
```
