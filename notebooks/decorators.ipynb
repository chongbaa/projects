```python
import time

# 装饰器函数
def timer_decorator(func):
    def wrapper():
        start_time = time.time()
        result = func()  # 执行原函数
        end_time = time.time()
        print(f"{func.__name__} 执行时间: {end_time - start_time:.4f}秒")
        return result
    return wrapper

# 使用装饰器
@timer_decorator
def say_hello():
    time.sleep(1)  # 模拟耗时操作
    print("Hello, World!")

# 使用装饰器
@timer_decorator  
def count_numbers():
    total = 0
    for i in range(1000000):
        total += i
    return total

# 测试
say_hello()
print("结果:", count_numbers())
```
