```python
# 1. 解包 tuple → 得到 list
a, *b = (1, 2, 3, 4)
print(type(b))  # <class 'list'>
print(a, b)

# 2. 解包 list → 得到 list
a, *b = [1, 2, 3, 4]
print(type(b))  # <class 'list'>
print(a, b)

# 3. 解包 str → 得到 list
a, *b = "hello"
print(type(b))  # <class 'list'>
print(a, b)

# 4. 解包 set → 得到 list
a, *b = {1, 2, 3, 4}
print(type(b))  # <class 'list'>
print(a, b)

# 5. 解包 dict（键）→ 得到 list
a, *b = {"x": 1, "y": 2, "z": 3}
print(type(b))  # <class 'list'>
print(a, b)

a, *b = {"x": 1, "y": 2, "z": 3}.values()
print(type(b))  # <class 'list'>
print(a, b)

a, *b = {"x": 1, "y": 2, "z": 3}.items()
print(type(a))  # <class 'tuple'>
print(type(b))  # <class 'list'>
print(a, b)
```
