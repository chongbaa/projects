```python
#[:] 数据集切片
items8 = ['apple', 'strawberry', 'pitaya', 'peach', 'watermelon']
print(items8[1:3:1])     # ['strawberry', 'pitaya']
print(items8[0:3:1])     # ['apple', 'strawberry', 'pitaya']
print(items8[0:5:2])     # ['apple', 'pitaya', 'watermelon']

print(items8[-4:-2:1])   # ['strawberry', 'pitaya']
print(items8[-2:-6:-1])  # ['peach', 'pitaya', 'strawberry', 'apple']

# 起始省略 → 从开头开始
# 结束省略 → 到结尾结束
# 步长省略 → 默认步长为1
t2 = ('骆昊', 45, True, '四川成都')
print(t2[:2])   # ('骆昊', 45)
print(t2[::3])  # ('骆昊', '四川成都')
```

```python
#占位符一位浮点表示法： ':.1f'
f = float(input('请输入华氏温度: '))
c = (f - 32) / 1.8
#print(f:.1f) SyntaxError: invalid decimal literal
print(f'{f:.1f}华氏度 = {c:.1f}摄氏度')
```

```python
# print((a = 10))
# SyntaxError: invalid syntax
# 海象运算符
print(a := 10)    # 10
print(a)          # 10
```

```python
#格式打印中，占位符里可用海象运算符
import math

radius = float(input('请输入圆的半径: '))  # 输入: 5.5
perimeter = 2 * math.pi * radius
area = math.pi * radius ** 2
print(f'{perimeter = :.2f}')  # 输出：perimeter = 34.56
print(f'{area = :.2f}')       # 输出：area = 95.03
```

```python
#格式化只是控制显示，不会改变实际数值精度。
#如果需要精确计算（如财务），应使用 Decimal。

from decimal import Decimal, ROUND_HALF_UP

num = Decimal("3.1415926")
result = float(num.quantize(Decimal("0.01"), rounding=ROUND_HALF_UP))
print(result)
```
